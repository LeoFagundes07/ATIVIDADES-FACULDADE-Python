```python
import matplotlib.pyplot as plt

# Passo 1: Definir a classe Livro
class Livro:
    def __init__(self, titulo, autor, genero, quantidade):
        self.titulo = titulo
        self.autor = autor
        self.genero = genero
        self.quantidade = quantidade

# Passo 2: Criar a lista de livros (nossa "biblioteca")
biblioteca = []

# --- Passo 3: Implementar funções para gerenciar os livros ---

# Função para cadastrar um novo livro [cite: 31]
def cadastrar_livro():
    print("\n--- Cadastro de Novo Livro ---")
    try:
        titulo = input("Digite o título do livro: ")
        autor = input("Digite o autor do livro: ")
        genero = input("Digite o gênero do livro: ")
        quantidade = int(input("Digite a quantidade disponível: "))

        novo_livro = Livro(titulo, autor, genero, quantidade)
        biblioteca.append(novo_livro)

        print(f"\n OK Livro '{titulo}' cadastrado com sucesso!")

    except ValueError:
        print("\nX Erro: A quantidade deve ser um número inteiro. Livro não cadastrado.")
    except Exception as e:
        print(f"\nX Ocorreu um erro inesperado: {e}")

# Função para listar todos os livros [cite: 32]
def listar_livros():
    print("\n--- Lista de Todos os Livros ---")
    if not biblioteca:
        print("A biblioteca está vazia. Nenhum livro cadastrado.")
        return

    for i, livro in enumerate(biblioteca):
        print(f"Livro {i+1}:")
        print(f"  Título: {livro.titulo}")
        print(f"  Autor: {livro.autor}")
        print(f"  Gênero: {livro.genero}")
        print(f"  Quantidade: {livro.quantidade}")
        print("-" * 20)

# Função para buscar um livro pelo título [cite: 34]
def buscar_livro():
    print("\n--- Busca de Livro por Título ---")
    if not biblioteca:
        print("A biblioteca está vazia. Impossível buscar.")
        return

    titulo_busca = input("Digite o título (ou parte dele) que deseja buscar: ")
    encontrados = []

    # Busca case-insensitive (ignora maiúsculas/minúsculas)
    for livro in biblioteca:
        if titulo_busca.lower() in livro.titulo.lower():
            encontrados.append(livro)

    if not encontrados:
        print(f"\nNenhum livro encontrado com o termo '{titulo_busca}'.")
    else:
        print(f"\n{len(encontrados)} livro(s) encontrado(s):")
        for livro in encontrados:
            print(f"  - Título: {livro.titulo}, Autor: {livro.autor}, Qtd: {livro.quantidade}")

# --- Passo 4: Utilizar a biblioteca Matplotlib para gerar um gráfico --- [cite: 35]
def gerar_grafico_generos():
    print("\n--- Gráfico de Livros por Gênero ---")
    if not biblioteca:
        print("A biblioteca está vazia. Cadastre livros para gerar o gráfico.")
        return

    # 1. Contar a quantidade de livros por gênero
    contagem_generos = {}
    for livro in biblioteca:
        # Usa .title() para padronizar (ex: "ficção" e "Ficção" contam juntos)
        genero_padronizado = livro.genero.strip().title()

        if genero_padronizado in contagem_generos:
            contagem_generos[genero_padronizado] += livro.quantidade
        else:
            contagem_generos[genero_padronizado] = livro.quantidade

    if not contagem_generos:
        print("Não foi possível gerar o gráfico (sem dados válidos).")
        return

    # 2. Preparar dados para o gráfico
    generos = list(contagem_generos.keys())
    quantidades = list(contagem_generos.values())

    # 3. Gerar o gráfico
    plt.figure(figsize=(10, 6)) # Define o tamanho da figura
    plt.bar(generos, quantidades, color='royalblue')
    plt.title('Quantidade de Livros por Gênero na Biblioteca')
    plt.xlabel('Gêneros Literários')
    plt.ylabel('Quantidade de Livros')
    plt.xticks(rotation=45, ha='right') # Rotaciona o nome dos gêneros se forem muitos
    plt.tight_layout() # Ajusta o layout para evitar cortes

    print("Exibindo o gráfico...")
    plt.show() # Exibe o gráfico


# --- Passo 5: Testar o sistema (Menu Principal) ---
def menu_principal():
    while True:
        print("\n=============================================")
        print("  Sistema de Gerenciamento da Biblioteca   ")
        print("=============================================")
        print("1. Cadastrar Novo Livro")
        print("2. Listar Todos os Livros")
        print("3. Buscar Livro por Título")
        print("4. Gerar Gráfico de Gêneros")
        print("5. Sair")
        print("---------------------------------------------")

        escolha = input("Escolha uma opção (1-5): ")

        if escolha == '1':
            cadastrar_livro()
        elif escolha == '2':
            listar_livros()
        elif escolha == '3':
            buscar_livro()
        elif escolha == '4':
            gerar_grafico_generos()
        elif escolha == '5':
            print("\nEncerrando o sistema. Até logo!")
            break
        else:
            print("\nX Opção inválida. Por favor, digite um número entre 1 e 5.")

# Executa o menu principal
menu_principal()
```

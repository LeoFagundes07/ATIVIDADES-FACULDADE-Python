```python
import os

# Lista para armazenar os pacientes
pacientes = []

def limpar_tela():
    os.system('cls' if os.name == 'nt' else 'clear')

def cadastrar_paciente():
    print("--- CADASTRAR PACIENTE ---")
    nome = input("Nome do paciente: ")
    while True:
        try:
            idade = int(input("Idade: "))
            break
        except ValueError:
            print("Digite um número válido.")
    telefone = input("Telefone: ")
    pacientes.append({"nome": nome, "idade": idade, "telefone": telefone})
    print("\nPaciente cadastrado!")
    input("Enter para continuar...")

def ver_estatisticas():
    print("--- ESTATÍSTICAS ---")
    total = len(pacientes)
    if total == 0:
        print("Nenhum dado ainda.")
    else:
        media = sum(p["idade"] for p in pacientes) / total
        mais_novo = min(pacientes, key=lambda p: p["idade"])
        mais_velho = max(pacientes, key=lambda p: p["idade"])
        print(f"Total: {total}")
        print(f"Média Idade: {media:.1f}")
        print(f"Mais novo: {mais_novo['nome']} ({mais_novo['idade']})")
        print(f"Mais velho: {mais_velho['nome']} ({mais_velho['idade']})")
    input("Enter para continuar...")

def buscar_paciente():
    busca = input("Nome para buscar: ").lower()
    encontrados = [p for p in pacientes if busca in p["nome"].lower()]
    if encontrados:
        for p in encontrados:
            print(f"{p['nome']} - {p['idade']} anos")
    else:
        print("Não encontrado.")
    input("Enter para continuar...")

def listar_todos():
    print("--- LISTA GERAL ---")
    for p in pacientes:
        print(f"{p['nome']} - {p['idade']} anos - {p['telefone']}")
    input("Enter para continuar...")

def menu():
    while True:
        limpar_tela()
        print("=== CLÍNICA VIDA+ ===")
        print("1. Cadastrar | 2. Estatísticas | 3. Buscar | 4. Listar | 5. Sair")
        op = input("Opção: ")
        if op == "1": cadastrar_paciente()
        elif op == "2": ver_estatisticas()
        elif op == "3": buscar_paciente()
        elif op == "4": listar_todos()
        elif op == "5": break

if __name__ == "__main__":
    menu()
```
